```python
import numpy as np
import pandas as pd
from math import radians, cos, sin, asin, sqrt
```

```python
stations=pd.read_csv('../../raw_data/stations_df_st.csv')
```

```python
stations_f=stations[['s_id','Station_name','s_lat','s_lon']]
stations_f
```

```python
def dist(lat1, long1, lat2, long2):
    """
Replicating the same formula as mentioned in Wiki
    """
    # convert decimal degrees to radians 
    lat1, long1, lat2, long2 = map(radians, [lat1, long1, lat2, long2])
    # haversine formula 
    dlon = long2 - long1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    # Radius of earth in kilometers is 6371
    km = 6371* c
    return km
```

```python
def find_nearest(lat, long):
    distances = stations_f.apply(
        lambda row: dist(lat, long, row['s_lat'], row['s_lon']), 
        axis=1)
    return stations_f.loc[distances.idxmin(), 's_id']
```
